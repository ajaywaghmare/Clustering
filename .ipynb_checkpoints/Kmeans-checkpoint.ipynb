```python
# Converting dataset from numpy array to pandas dataframe
y = pd.DataFrame(data=iris.target, 
                 index=np.array(range(1, 151)),
                 columns=np.array(range(1, 2)))
```

```python
# Convering target class into numerical form
y[1] = y[1].replace(['Iris-setosa',
                     'Iris-versicolor','Iris-virginica'], [0,1,2])
```

```python
plt.scatter(x.iloc[:,0],x.iloc[:,1], c = y.iloc[:,0], cmap='gist_rainbow')
plt.xlabel('Spea1 Length', fontsize=18)
plt.ylabel('Sepal Width', fontsize=18)
plt.show()
```

```python
kmeans_kwargs = {"n_init": 10,"max_iter": 300,
                 "random_state": 2 }
```

```python
silhouette_coefficients = []

for k in range(2, 11):
    km = KMeans(n_clusters=k, **kmeans_kwargs)
    km.fit(x)
    silhouette_s = silhouette_score(x, km.labels_)
    silhouette_coefficients.append(silhouette_s)
```

```python
silhouette_coefficients
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 11), silhouette_coefficients)
plt.xticks(range(2, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")
plt.show()
```

```python
km = KMeans(n_clusters=3, **kmeans_kwargs)
km.fit(x)
```

```python
km.labels_
```

```python
pred_result = pd.DataFrame(data=km.labels_,
          index=np.array(range(1, 151)),
          columns=np.array(range(1, 2)))
```

```python
adjusted_rand_score(y.iloc[:,0], km.labels_)
```

```python
f1_sc = f1_score(y, pred_result, average = 'macro')
print(f1_sc)
```

```python
km.cluster_centers_
```

```python
km.inertia_
```
